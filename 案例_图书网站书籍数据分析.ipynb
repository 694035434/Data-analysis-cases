```python
# 拉格朗日插值法
from scipy.interpolate import lagrange #拉格朗日插值函数

data=pd.read_excel("data/lagr_data.xls")
data
# lagrange(列向量，被插值的位置，取前后的数据个数)
def d_lagrange(c,n,k=5):
    new=pd.concat([c[n-k:n],c[n+1:n+k+1]])
#     print(new.index)
#     print(new.values)
    a=lagrange(new.index,new.values)(n)
    return a
for i in range(len(data["销量"])):
    if (data["销量"].isnull())[i]==True:
        data["销量"][i]=d_lagrange(data["销量"],i)
data
```
